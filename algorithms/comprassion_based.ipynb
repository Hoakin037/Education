color sort

```python
def color_sort(arr):
    low = 0          # граница для 0
    mid = 0          # текущий рассматриваемый элемент
    high = len(arr) - 1  # граница для 2

    while mid <= high:
        if arr[mid] == 0:
            arr[low], arr[mid] = arr[mid], arr[low]
            low += 1
            mid += 1
        elif arr[mid] == 1:
            mid += 1
        else:  # arr[mid] == 2
            arr[high], arr[mid] = arr[mid], arr[high]
            high -= 1
lst = [1,1,2,0,1,2,0,0,2,1]
color_sort(lst)
lst
```

bubble sort

```python
def bubble_sort(self, lst) -> None:
        """
        Mutates lst so that it is sorted via swapping adjacent elements until
        the entire lst is sorted.
        """
        has_swapped = True
        # if no swap occurred, lst is sorted
        while has_swapped:
            has_swapped = False
            for i in range(len(lst) - 1):
                if lst[i] > lst[i + 1]:
                    # Swap adjacent elements
                    lst[i], lst[i + 1] = lst[i + 1], lst[i]
                    has_swapped = True
```

```python
def heightChecker(heights):
    counter = 0
    sort_h = sorted(heights)
    for i in range(len(heights)):
        if heights[i] != sort_h[i]:
            counter += 1
    return counter
```

```python
def insertion_sort(lst):
    """
    Mutates elements in lst by inserting out of place elements into appropriate
    index repeatedly until lst is sorted
    """
    for i in range(1, len(lst)):
        current_index = i

        while current_index > 0 and lst[current_index - 1] > lst[current_index]:
            # Swap elements that are out of order
            
            lst[current_index], lst[current_index - 1] = lst[current_index - 1], lst[current_index]
            current_index -= 1

        

lst = [2,1,6,8,2,9,3,0,1]
insertion_sort(lst)
lst
```

```python
from typing import Optional

def insertionSortList(lst):
    head = 0
    next = 1
    for i in range(len(lst) - 1):
        if lst[head] > lst[next]:
            j = 0
            while lst[head] > lst[next]:
                sort_list = lst[:next]
                if lst[next] < sort_list[j]:
                    sort_list[j] = lst[next]
                    sort_list.append(lst[next])
                    lst.pop(head)
                j += 1
            head += 1
            next += 1
    return sort_list

head = [4,2,1,3]
insertion_sort(head)
head


class ListNode:
     def __init__(self, val=0, next=None):
         self.val = val # значение в узле
         self.next = next # ссылка на следующий узел (или None)

class Solution:
    def insertionSortList(self, head: Optional[ListNode]) -> Optional[ListNode]:
        if not head or not head.next:
            return head  # список из 0 или 1 элемента уже отсортирован
       
        # Создаём фиктивный узел (dummy), чтобы упростить вставку в начало
        dummy = ListNode(0)
        dummy.next = head  # dummy → 3 → 1 → 4 → 2 → None
       
        # sorted_tail — указатель на последний узел уже отсортированной части
        sorted_tail = head  # изначально отсортирована только первая нода: [3]
        curr = head.next    # текущий узел, который мы будем вставлять
       
        while curr:  # пока есть что вставлять
            if sorted_tail.val <= curr.val:
                # Случай 1: текущий элемент больше последнего в отсортированной части
                # → он и так на своём месте в конце
                # Просто расширяем отсортированную часть
                sorted_tail = curr
                curr = curr.next
            else:
                # Случай 2: curr.val < sorted_tail.val
                # → нужно найти место в уже отсортированной части (от dummy до sorted_tail)
                # и вставить curr туда
                
                prev = dummy  # начинаем с dummy
                # Идём по отсортированной части, пока не найдём место:
                # ищем первый узел, который >= curr.val
                while prev.next.val < curr.val:
                    prev = prev.next
               
                # Теперь prev.next — это первый узел больше или равный curr.val
                # Вставляем curr между prev и prev.next

                next_node = curr.next        # сохраняем, куда curr смотрел
                curr.next = prev.next        # curr теперь смотрит на тот, что был после prev
                prev.next = curr             # prev теперь смотрит на curr
               
                # Отцепляем curr от неотсортированной части
                sorted_tail.next = next_node  # бывший "хвост" теперь смотрит на следующий после curr
                curr = next_node               # переходим к следующему кандидату на вставку
        
        return dummy.next  # настоящий начало списка (без dummy)
```

Односвязный список — это цепочка объектов ListNode, где каждый знает только про следующий. head — это первый узел списка.
