### <font color='middleblue'>How to use *args and **kwargs</font>

```python
def print_wishlist(name, *presents):
    print(f"{name} wants on birthday: ", end='')
    for gift in presents:
        print(gift, end=", ")

print_wishlist("Jason", 'car', 'Samsung S24 Ultra', 'apple watch')
```

```python
def print_wishlist(name, *presents):
    print(f"{name} wants on birthday: ", end='')
    print(*presents, sep=", ")

print_wishlist("Jason", 'car', 'Samsung S24 Ultra', 'apple watch')
```

```python
def print_wishlist(**person_list):
    for person, gifts in person_list.items():
        print(f"{person} wants on birthday: {gifts}")

print_wishlist(Lara='flowers', Nick='book', Karren='car')
```

```python
def print_wishlist(**person_list):
    for person, gifts in person_list.items():
        list = ''.join(f"{i}, " for i in gifts)
        print(f"{person} wants on birthday: {list}")

print_wishlist(Lara=['flowers', 'cakes'], Nick=['book', 'new orchidea'], Karren=['car', 'Apple 12 pro max'])
```
