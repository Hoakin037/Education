### <font color='middleblue'>Closing functions</font>

```python
def outer_func(x: int):
    def inner_func():
        return x + 2
    
    return inner_func

func = outer_func(2)
func()
    
```

```python
def outer_func():
    x = 2
    def inner_func():
        return x + 2
    
    return inner_func

func = outer_func()
func()
```

```python
def outer_func(x):
    def inner_func(y):
        return x + y
    
    return inner_func

func = outer_func(2)
func(2)
```

```python
def count_calls():
    counter = 0

    def closure(print_result=False):
        nonlocal counter
        if print_result:
            return counter
        counter += 1
        return counter
    return closure


counter = count_calls()     # Вызвав функцию, получаем счетчик (замыкание)

for _ in range(5):
    counter()               # Вызываем счетчик

print(counter(True))        # Проверяем результат подсчета: 5

for _ in range(2):
    counter()

print(counter(1)) # То же самое, что и counter(True)
```

```python
def use_state(iv):
    val = iv
    def getval():
        return val
    def setval(newval):
        nonlocal val 
        val = newval
        return val
    return [getval, setval]    

[getter, setter] = use_state(1)
print(getter())
print(setter(5))
print(getter())
```
