### <font color='middleblue'>Decorators</font>

```python
def print_hello_world(func):
    print("Hello, ", end="")
    return func

@print_hello_world
def print_world():
    print("World!")

print_world()
```

```python
def add_points(func):
    def adder():
        return '...' + func() + '...'
    return adder

def add_under_scores(func):
    def adder():
        return '__' + func() + '__'
    return adder

@add_points
@add_under_scores
def print_line():
    return "hello"

print(print_line())
```
