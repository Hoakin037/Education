### <font color="middleblue">Lists, tuples and sets. When we should use each of them</font>

#### Lists

Изменяемые массивы данных, могут содержать как однотипные, так и разнотипные данные, могут быть многомерными. 

```python
l1 = [1,2,3,4]
l2 = [1, 'a', [[0], [1]], 4.0]
l3 = [i for i in 'advbsbass']
l4 = "gw   ge w  i j o i j m a d a".split()
print(l1,l2,l3,l4, sep="\n")

l1.append(5)
l2.extend(l3)
l3[:] = [0]*len(l3)
l4.pop(0)
print(l1,l2,l3,l4, sep="\n")
```

#### Tuples
Кортежи — это неизменяемые последовательности, которые хранят элементы по индексу. Их можно использовать в качестве ключей в словарях. Для поиска элемента в кортеже Python приходится последовательно проверять каждый элемент, пока не будет найдено совпадение, что занимает линейное время, в среднем O(n). Из-за своей неизменяемости кортежи занимают меньше памяти и быстрее в создании, чем списки, но это не сказывается на скорости поиска элементов так же эффективно, как у сетов. 

```python
t1 = (1,2,3,4,5)
t2 = ('a', 'b', 'c', 'd')
t3 = (1, 'a')
print(t1,t2,t3, sep="\n")
```

#### Sets
Сеты изменяемый тип данных данных в питоне, хранит только уникальные значения, но они могут иметь разные типы. Могут содержать числа (int, float), строки (str), кортежи (tuple) и другие неизменяемые объекты. Однако они не могут содержать изменяемые типы данных, такие как списки (list), словари (dict) или другие множества (set), потому что элементы множества должны быть хешируемыми, а изменяемые объекты — нет. 
Сеты используют хеш-таблицу, что позволяет выполнять операции поиска, добавления и удаления элементов очень быстро, в среднем за константное время O(1), независимо от размера сета. При этом элементы сета должны быть хешируемыми, то есть иметь неизменяемый тип данных (например, строки, числа), и порядок элементов не сохраняется. 
С множествами можно выполнять множество операций: находить объединение, пересечение...

len(s) - число элементов в множестве (размер множества).
x in s - принадлежит ли x множеству s.
set.isdisjoint(other) - истина, если set и other не имеют общих элементов.
set == other - все элементы set принадлежат other, все элементы other принадлежат set.
set.issubset(other) или set <= other - все элементы set принадлежат other.
set.issuperset(other) или set >= other - аналогично.
set.union(other, ...) или set | other | ... - объединение нескольких множеств.
set.intersection(other, ...) или set & other & ... - пересечение.
set.difference(other, ...) или set - other - ... - множество из всех элементов set, не принадлежащие ни одному из other.
set.symmetric_difference(other); set ^ other - множество из элементов, встречающихся в одном множестве, но не встречающиеся в обоих.
set.copy() - копия множества.
И операции, непосредственно изменяющие множество:

set.update(other, ...); set |= other | ... - объединение.
set.intersection_update(other, ...); set &= other & ... - пересечение.
set.difference_update(other, ...); set -= other | ... - вычитание.
set.symmetric_difference_update(other); set ^= other - множество из элементов, встречающихся в одном множестве, но не встречающиеся в обоих.
set.add(elem) - добавляет элемент в множество.
set.remove(elem) - удаляет элемент из множества. KeyError, если такого элемента не существует.
set.discard(elem) - удаляет элемент, если он находится в множестве.
set.pop() - удаляет первый элемент из множества. Так как множества не упорядочены, нельзя точно сказать, какой элемент будет первым.
set.clear() - очистка множества.

```python
s1 = {1,2,3,4,5,6}
s2 = {1,1,1,2,3,4,5,6,6,7}
s3 = set().difference_update(s2)

print(s1,s2,s3, sep="\n")
```
